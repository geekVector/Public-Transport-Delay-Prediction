```python
import pandas as pd
```

```python
df=pd.read_csv('public_transport_delays.csv')
```

```python
df
```

```python
df['transport_type']=df['transport_type'].str.lower()
```

```python
df['origin_station']=df['origin_station'].str.lower()
df['origin_station']=df['origin_station'].str.replace('_',' ')
```

```python
df['destination_station']=df['destination_station'].str.lower()
df['destination_station']=df['destination_station'].str.replace('_',' ')
```

```python
df['scheduled_departure']=df['scheduled_departure'].str.slice(0,5)
```

```python
df['scheduled_arrival']=df['scheduled_arrival'].str.slice(0,5)
```

```python
df['event_type']=df['event_type'].fillna('no event')
```

```python
df['holiday']=df['holiday'].map({0:'no',1:'yes','0':'no','1':'yes'})
```

```python
df['peak_hour']=df['peak_hour'].map({0:'no',1:'yes','0':'no','1':'yes'})
```

```python
df['season']=df['season'].str.lower()
```

```python
df['delayed']=df['delayed'].map({0:'no',1:"yes",'0':'no','1':'yes'})
```

```python
df['weather_condition'] = df['weather_condition'].astype(str).str.lower()
```

```python
df['weather_condition'].dtype
```

```python
df1=df.drop(['trip_id','date','time','route_id','actual_departure_delay_min','actual_departure_delay_min','actual_arrival_delay_min','humidity_percent','wind_speed_kmh','precipitation_mm','weekday'],axis=1)
```

```python
df1.head()
```

```python
print(df1.isnull().sum())
```

```python
x=df1.drop(['delayed'],axis=1)
y=df1['delayed']
```

```python
num = []
cat = []

for col in df1:
    if df1[col].dtype == 'object':
        cat.append(col)    # Fixed: Changed [] to ()
    elif df1[col].dtype == 'int64' or df1[col].dtype=='float64':
        num.append(col)    # This one was already correct!
```

```python
num
```

```python
cat.remove('delayed')
```

```python
cat
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder,OrdinalEncoder,OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
preprocessing=ColumnTransformer([
    ('num',StandardScaler(),num),
    ('char',OneHotEncoder(handle_unknown='ignore',sparse_output=False),cat)
])
```

```python
preprocessing
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
```

```python
pipeline=Pipeline([
    ('preprocessor',preprocessing),
    ('model',RandomForestClassifier(n_estimators=100,
    max_depth=10,
    random_state=42,
    class_weight='balanced'))
]).fit(x_train,y_train)
```

```python
y_pred=pipeline.predict(x_test)
```

```python
y_pred
```

```python
y_train.value_counts()
```

```python
y_test
```

```python
from sklearn.metrics import f1_score
```

```python
f1=f1_score(y_test, y_pred, pos_label='yes')
print(f1)
```
